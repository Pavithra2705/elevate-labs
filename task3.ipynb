```python
# =======================
# EDA on Iris Dataset
# =======================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
df["species"] = df["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})

# -----------------------
# 1. Histograms (Numerical Features)
# -----------------------
df.hist(figsize=(10, 8))
plt.suptitle("Distribution of Numerical Features")
plt.show()

# -----------------------
# 2. Count Plot (Categorical Feature)
# -----------------------
sns.countplot(x="species", data=df)
plt.title("Count of Iris Species")
plt.show()

# -----------------------
# 3. Box Plot (Outlier Detection)
# -----------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop("species", axis=1))
plt.title("Box Plot for Outlier Detection")
plt.show()

# -----------------------
# 4. Correlation Heatmap
# -----------------------
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop("species", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# -----------------------
# 5. Summary Insights
# -----------------------
print("""
INSIGHTS:
- Dataset is clean with no missing values
- Classes are balanced
- Petal length and petal width are highly correlated
- Petal features are most important for prediction
- Minimal outliers detected
""")
```
